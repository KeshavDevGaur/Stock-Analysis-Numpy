## 📌 Introduction

This project performs a comprehensive analysis of stock market data using only NumPy, without relying on additional libraries.

## 📊 Dataset Description

The dataset consists of stock prices for multiple stocks over several days.

- **Rows** represent different stocks  
- **Columns** represent daily prices  

### Example:
- Stock A → prices over 5 days  
- Stock B → prices over 5 days  
- Stock C → prices over 5 days  

```python
import numpy as np
```

```python
stocks = np.array(["A", "B", "C"])

prices = np.array([
    [100, 102, 101, 105, 110],   
    [200, 198, 202, 205, 207],  
    [50, 55, 53, 52, 58]         
])

print("Stock Prices:\n", prices)
```

```python
daily_returns = (prices[:, 1:] - prices[:, :-1]) / prices[:, :-1] * 100

print("📊 Daily Returns (%):\n", daily_returns)
```

```python
avg_return = np.mean(daily_returns, axis=1)

for i in range(len(stocks)):
    print(f"Average Return of Stock {stocks[i]}: {avg_return[i]:.2f}%")
```

```python
volatility = np.std(daily_returns, axis=1)

for i in range(len(stocks)):
    print(f"Volatility of Stock {stocks[i]}: {volatility[i]:.2f}")
```

```python
total_growth = ((prices[:, -1] - prices[:, 0]) / prices[:, 0]) * 100

best_stock = np.argmax(total_growth)
worst_stock = np.argmin(total_growth)

print(f"🚀 Best Stock: {stocks[best_stock]} ({total_growth[best_stock]:.2f}%)")
print(f"⚠️ Worst Stock: {stocks[worst_stock]} ({total_growth[worst_stock]:.2f}%)")
```

```python
max_gain = np.max(daily_returns)
index = np.unravel_index(np.argmax(daily_returns), daily_returns.shape)

print(f"📈 Highest Gain: {max_gain:.2f}%")
print(f"Stock: {stocks[index[0]]}, Day: {index[1] + 1}")
```

```python
window = 3

moving_avg = np.array([
    np.convolve(stock, np.ones(window)/window, mode='valid')
    for stock in prices
])

print("📊 Moving Averages:\n", moving_avg)
```

```python
cumulative_returns = (prices / prices[:, [0]]) - 1

print("📈 Cumulative Returns:\n", cumulative_returns)
```

```python
portfolio = np.mean(prices, axis=0)

portfolio_return = ((portfolio[-1] - portfolio[0]) / portfolio[0]) * 100

print(f"💼 Portfolio Return: {portfolio_return:.2f}%")
```

```python
predicted_next = np.mean(prices[:, -3:], axis=1)

for i in range(len(stocks)):
    print(f"🔮 Predicted Next Price of {stocks[i]}: {predicted_next[i]:.2f}")
```

```python
print("📊 Risk vs Return:")

for i in range(len(stocks)):
    print(f"Stock {stocks[i]} → Return: {avg_return[i]:.2f}% | Risk: {volatility[i]:.2f}")
```

```python
correlation_matrix = np.corrcoef(prices)

print("📊 Correlation Matrix:\n", correlation_matrix)
```

```python
risk_free_rate = 1  

sharpe_ratio = (avg_return - risk_free_rate) / volatility

print("📈 Sharpe Ratio:")

for i in range(len(stocks)):
    print(f"Stock {stocks[i]}: {sharpe_ratio[i]:.2f}")
```

```python
drawdown = []

for stock in prices:
    peak = np.maximum.accumulate(stock)
    dd = (stock - peak) / peak
    drawdown.append(np.min(dd))

drawdown = np.array(drawdown)

print("📉 Maximum Drawdown:")

for i in range(len(stocks)):
    print(f"Stock {stocks[i]}: {drawdown[i]:.2f}")
```

```python
mean_price = np.mean(prices, axis=1, keepdims=True)
std_price = np.std(prices, axis=1, keepdims=True)

z_scores = (prices - mean_price) / std_price

print("⚠️ Z-Scores:\n", z_scores)
```

```python
weights = np.array([0.5, 0.3, 0.2])

portfolio_weighted = np.dot(weights, prices)

portfolio_return = ((portfolio_weighted[-1] - portfolio_weighted[0]) / portfolio_weighted[0]) * 100

print(f"💼 Weighted Portfolio Return: {portfolio_return:.2f}%")
```
